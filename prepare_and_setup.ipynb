## Import libraries

## Read data

```python
input_path = '../../DATA/AIHW-People-with-disability-in-Australia-2018/generated_data/'
input_file = 'transformed_data.csv'
```

```python
df = pd.read_csv(input_path + input_file, sep=',', encoding='utf-8')
df
```

The transformed data contains estimated number of people with a disability by employment status as well as the proportion. 

## Visualize data

Here, we'd like to visualize the generated data. Below code demonstrates how to visualize the data by using Matplot library.

**Matplot** : Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python.

```python
from treelib import Node, Tree

tree = Tree()

tree.create_node("2,057,500 people aged 15-64 with disability", "root")  # No parent means its the root node
tree.create_node("In the labour force (53.4%)",  "In the labour force"   , parent="root")
tree.create_node("Not in the labour force (46.6%)",  "Not in the labour force"   , parent="root")
tree.create_node("Employed (47.8%)",  "employed"   , parent="In the labour force")
tree.create_node("Unemployed (5.5%)",  "unemployed"   , parent="In the labour force")
tree.create_node("Working full-time (28.3%)",  "full-time"   , parent="employed")
tree.create_node("Working part-time (19.6%)",  "part-time"   , parent="employed")
tree.save2file('img/disability participant to labour force.txt')
tree.show()
```

The above tree graph shows hierarchy structure within employment status of people aged 15-64 with disability. In the total 2,057,500 people aged 15-64 with disability people, 53.4% are belong to labour force while the other 46.6% are not in the labour force. In the 53.4% labour force, 47.8% are employed where 28.3% are full-time and 19.6% are part-time.

## Insight

**53.4% of working-age (aged 15-64) people with disability participated in the Australian labour force in 2018.**

```python
value = float(df['proportion(%)'].iloc[0])
print('{}% of working-age (aged 15-64) people with disability participated in the Australian labour force in 2018.'.format(round(value, 1)))
```
